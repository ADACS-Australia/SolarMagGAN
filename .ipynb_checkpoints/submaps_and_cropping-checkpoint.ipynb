
# Cropping a Map


How to crop a map by using submap.


```python
import astropy.units as u
from astropy.coordinates import SkyCoord

import sunpy.map
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle
```

We start with the sample data



```python
swap_map = sunpy.map.Map("/home/cam/Documents/winter_project/SolarMagGAN/FITS_DATA/Dopplergrams/hmi.v_45s.20150101_000045_TAI.2.Dopplergram.fits")
```

To crop the data you create a submap, specifying the top right and bottom
left as SkyCoord objects.



```python
coords = SkyCoord(106.4*u.deg, 18.1*u.deg, frame= "heliographic_carrington")
coords = swap_map.world_to_pixel(coords)
x = np.floor(coords.x.value)
y = np.floor(coords.y.value)
x,y
```

```python
fig = plt.figure()
# Provide the Map as a projection, which creates a WCSAxes object
ax = plt.subplot(projection=swap_map)

im = swap_map.plot()

swap_map.plot()

# Prevent the image from being re-scaled while overplotting.
ax.set_autoscale_on(False)

swap_map.plot()

w = h = 300.
r = Rectangle((x - w/2,y - h/2), w, h, edgecolor='red', facecolor='none')
p = ax.add_patch(r)

plt.savefig('test.png', dpi = 600)
plt.show()
```
